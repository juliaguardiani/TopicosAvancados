##1.a
Escreva um programa em Python que utilizando as bibliotecas numpy e matplotlib, execute as seguintes tarefas:

- Gere uma distribuição aleatória de dados para X e Y;
- Plote uma contra a outra, construindo um gráfico de dispersão do tipo Scatter
Plot com o marcador ‘+’, na cor vermelha;
- Salvar a figura no seu google drive;

```python
#Importações para o exercício #1.a.
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
import random #random.shuffle(sequencia,função). Sequencia: pode ser uma lista, tupla, string. Funcao~opcionao: o nome da função.
from googleapiclient.http import MediaFileUpload
from google.colab import files
```

```python
#Criar uma lista com números entre 1 a 500.
xa = list(range(100,500))
ya = list(range(100,500))

#Embaralha a lista de números x e y.
random.shuffle(xa)
random.shuffle(ya)
print("x=",xa)
print("y=",ya)

#Conf do gráfico.
plt.title('Distribuição aleatória')
plt.xlabel('"X" aleatório')
plt.ylabel('"Y" aleatório')

plt.scatter(xa,ya,color="red",marker="+") #Produzindo gráfico de dispersão.
fig1 = plt.gcf() #Salvando gráfico em uma variável.
plt.show()
#Salvando gráfico no drive.
caminho = '/content/gdrive/MyDrive/2021.1/TÓPICOS AVANÇADOS EM INFORMÁTICA II/'
fig1.savefig(caminho+'scat.png', dpi=100)
```

##1.b
- Plote duas senóides de cores diferentes em um mesmo gráfico, com amplitude e
frequência definidas pelo usuário;
- Salvar a figura no seu google drive;


```python
#Importações para o exercício #1.b.
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Frequência e amplitude definidas pelo usuário
frequencia1= input('Frequência para a primeira senoide?')
amplitude1 = input('Amplitude para a primeira senoide?')

frequencia1 = int(frequencia1)
amplitude1 = int(amplitude1)

time1 = np.arange(0,10,0.1)
calAmplitude1 = amplitude1*np.sin(frequencia1*time1)

frequencia2= input('Frequência para a segunda senoide?')
amplitude2 = input('Amplitude para a segunda senoide?')

frequencia2 = int(frequencia2)
amplitude2 = int(amplitude2)

time2 = np.arange(0,10,0.1)
calAmplitude2 = amplitude2*np.sin(frequencia2*time2)

#Conf do gráfico.
fig2 = plt.gcf() #Salvando gráfico em uma variável.
plt.plot(time1,calAmplitude1)
plt.plot(time2,calAmplitude2)

plt.title("Senóides")
plt.xlabel('tempo')
plt.ylabel('amplitude')

plt.show(fig2)
#Salvando gráfico no drive.
caminho = '/content/gdrive/MyDrive/2021.1/TÓPICOS AVANÇADOS EM INFORMÁTICA II/'
fig2.savefig(caminho+'scat.png', dpi=100)
```
